```python
#dropping irrelavant columns
df.drop(['identifierHash','type', 'seniorityAsMonths' , 'civilityTitle'] , axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
#dropping null values
df.dropna(inplace=True)
```

```python
df['hasProfilePicture'] = df['hasProfilePicture'].astype('int')
df['hasProfilePicture'].dtypes
```

```python
df.columns
```

```python
#renaming the coloumns
df.rename(columns={'socialNbFollowers':'Followers' , 'socialNbFollows':'Following'})
```

```python
# describe() method returns description of the data in the Dataframe.
df.describe()
```

```python
df[['productsListed','productsSold','productsPassRate','productsBought']].describe()
```

```python
# plotting a bar chart for gender and its count

ax= sns.countplot(x='gender',data=df)

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs product bought 

sales_gen = df.groupby(['gender'] , as_index=False)['productsBought'].sum().sort_values(by='productsBought',ascending=False)

sns.barplot(x = 'gender' , y = 'productsBought' , data=sales_gen)
```

```python
ax = sns.countplot(data=df  , x = 'seniority' , hue='gender')

for bars in  ax.containers :
    ax.bar_label(bars)
```

```python
# productsBought vs seniority

sales_sen = df.groupby(['seniority'] , as_index=False)['productsBought'].sum().sort_values(by='productsBought',ascending=False)

sns.barplot(x = 'seniority' , y= 'productsBought' ,data = sales_sen)
```

```python
ax = sns.countplot(data=df , x = 'language')

sns.set(rc={'figure.figsize': (7,5)})
for bars in ax.containers :
    ax.bar_label(bars)
```

```python
# language vs products bought

sales_1 = df.groupby(['language'] , as_index=False)['productsBought'].sum().sort_values(by='productsBought', ascending=False)

sns.barplot(x='language' ,y = 'productsBought', data=sales_1)
```

```python
# language vs products sold

sales_2 = df.groupby(['language'] , as_index=False)['productsSold'].sum().sort_values(by='productsSold' , ascending=False)

sns.barplot(x = 'language' , y = 'productsSold', data = sales_2)
```

```python
#Android users

ax = sns.countplot(data=df, x = 'hasAndroidApp' )

for bars in ax.containers :
    ax.bar_label(bars)
    
    

    
```
